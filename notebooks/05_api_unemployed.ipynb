```python
import requests
import pandas as pd 
from io import StringIO
```

```python
url = "https://portaldades.ajuntament.barcelona.cat/services/backend/rest/statistic/export?id=swx6ul5nve&fileformat=CSV"
headers = {"X-IBM-Client-Id": "f943a426a663183091a1401f55eec60d"}

response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    data = StringIO(response.content.decode('latin1'))
    df_unemployed = pd.read_csv(data, sep=',')  # Sep virgül
    print("✅ Veri çekildi")
    print(df_unemployed.head())
else:
    print(f"❌ HATA - Status Code: {response.status_code}")
```

```python
df_unemployed.head()
```

```python
df_unemployed = df_unemployed.rename(columns={
    'Dim-00:TEMPS': 'year',
    'Dim-01:TERRITORI': 'territory',
    'Dim-01:TERRITORI (order)': 'territory_order',
    'Dim-01:TERRITORI (type)': 'territory_type',
    'VALUE': 'num_unemployed'
})
```

```python
df_unemployed.tail(20)
```

```python
import numpy as np
for column in df_unemployed:
    unique_vals = np.unique(df_unemployed[column].fillna('0'))
    nr_values = len(unique_vals)
    if nr_values <= 85:
        print("The number of values for feature {} :{} -- {}".format(column, nr_values, unique_vals))
    else:
        print("The number of values for feature {} :{}".format(column, nr_values))
```

```python
df_unemployed['territory'] = df_unemployed['territory'].str.encode('latin1').str.decode('utf-8')
df_unemployed['territory_type'] = df_unemployed['territory_type'].str.encode('latin1').str.decode('utf-8')
```

```python
print(df_unemployed['territory'].unique()[:10])
print(df_unemployed['territory_type'].unique())
```

```python
print(sorted(df_unemployed['year'].unique()))
```

```python
df_unemployed['year'] = pd.to_datetime(df_unemployed['year'])
df_unemployed['year_quarter'] = df_unemployed['year'].dt.to_period('Q').astype(str)
```
